```python
import matplotlib.pyplot as plt
import numpy as np

# Create some data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create a figure
plt.figure(figsize=(12, 5))

# First subplot for the sine wave
plt.subplot(1, 2, 1)  # (rows, columns, panel number)
plt.plot(x, y1, 'r-')
plt.title('Sine Wave')
plt.xlabel('X')
plt.ylabel('sin(X)')

# Second subplot for the cosine wave
plt.subplot(1, 2, 2)
plt.plot(x, y2, 'b-')
plt.title('Cosine Wave')
plt.xlabel('X')
plt.ylabel('cos(X)')

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
